```python
import matplotlib.pyplot as plt
import csv
import os
import pandas as pd
import statistics as st
import numpy as np

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis,y_axis,label='Boston',marker="D", color="green", linewidth=2)
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()

# Add a grid.
plt.grid()


```

```python
#plot using Object-oriented method
fig,ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#Alt way of plot using Object-oriented method
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.bar(x_axis,y_axis,label='Boston',color='green')
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title('PyBer Fare by Month')

# Add the legend.
plt.legend()

```

```python
#Generate horizontal bar chart using MATLAB method
plt.barh(x_axis,y_axis,label='Boston',color='magenta')
plt.xlabel('Fare($)')
plt.ylabel('Date')

# Create a title
plt.title('PyBer Fare by Month')

# Add a legend
plt.legend()

#invert axis and show
plt.gca().invert_yaxis()
plt.show()
```

```python
#Generate horizontal bar chart using object oriented method
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis,color='cyan')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend(['Chicago'])

#invert data series
ax.invert_yaxis()
```

```python
#Scatter plot using MATLAB method and adding addl attribute to plot
#plt.plot(x_axis,y_axis,"o")

#Scatter plot using MATLAB method and scatter
plt.scatter(y_axis,x_axis,c="red")

plt.title("PyBer Fare by month")
plt.ylabel('Date')
plt.xlabel('Fare($)')

plt.gca().invert_yaxis()
plt.legend('Chicago')
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Scatter plor with object-oriented code method
fig,ax=plt.subplots()

#add attributes for color, edge color, transparency, line width
ax.scatter(y_axis,x_axis,s=[5*i for i in  y_axis],color='skyblue',edgecolor='black',alpha=0.8,linewidth=2)
ax.set_title("PyBer Fare by Month")

# Set the x limit between 0 and 50
ax.set_xlim(0, 50)

#Invert axis
ax.invert_yaxis()

```

```python
#pie chart using MATLAB method

#define colors in a list
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#define which label to explode and by how much
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#set fig frame size
plt.figure(figsize=(8,8))

#pie chart
plt.pie(y_axis,explode=explode_values,autopct='%.1f%%',labels=x_axis,colors=colors)
plt.show()
```

```python
#pie chart using OO method
fig,ax=plt.subplots(figsize=(8,8))

#define which label to explode and by how much
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#add shadow
ax.pie(y_axis,labels=x_axis,colors=colors,explode=explode_values,autopct='%.1f%%',shadow='true',startangle=-95)

ax.invert_yaxis()
plt.show()
```
